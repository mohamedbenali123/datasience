```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
X.shape
```

```python
# Trac√© du nuage de points initial
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Nuage de points avant le groupement des iris')
plt.show()
```

```python
# Cr√©ation d'un objet KMeans avec k clusters
k=3
kmeans = KMeans(n_clusters=k)

# Entra√Ænement du mod√®le
kmeans.fit(X)

# Obtention des affectations de cluster pour chaque √©chantillon
labels = kmeans.labels_

# Obtention des centres de cluster
centers = kmeans.cluster_centers_

#labels.shape
centers
```

```python
# Trac√© du nuage de points
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Nuage de points avec K-means ({k} clusters)')
plt.show()
```

```python
# Trac√© du nuage de points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Nuage de points avec K-means ({k} clusters)')
plt.show()
```

```python
# Trac√© du nuage de points avant le groupement des iris
plt.subplot(1, 2, 1)  # Sp√©cifier la position du subplot
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Nuage de points avec K-means ({k} clusters)')

# Trac√© du nuage de points avec K-means (k clusters)
plt.subplot(1, 2, 2)  # Sp√©cifier la position du subplot
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Nuage de points avec K-means ({k} clusters)')

plt.tight_layout()  # Pour am√©liorer l'espacement entre les sous-graphiques
plt.show()
```

```python
# Liste pour stocker les valeurs d'inertie
inertie = []

# Nombre de clusters √† tester
k_values = range(1, 15)

# Calcul de l'inertie pour chaque valeur de k
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    inertie.append(kmeans.inertia_)

# Trac√© de la courbe du coude
plt.plot(k_values, inertie, 'bx-')
plt.xlabel('Nombre de clusters (k)')
plt.ylabel('Inertie')
plt.title('Courbe du coude')
plt.show()
```

```python
# Calcul de l'inertie pour k=3
kmeans_3 = KMeans(n_clusters=3)
kmeans_3.fit(X)
inertie_3 = kmeans_3.inertia_
print("Inertie pour k=3 :", inertie_3)

# Calcul de l'inertie pour k=4
kmeans_4 = KMeans(n_clusters=4)
kmeans_4.fit(X)
inertie_4 = kmeans_4.inertia_
print("Inertie pour k=4 :", inertie_4)
```
