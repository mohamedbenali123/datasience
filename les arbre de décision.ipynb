```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
X=iris.data
y=iris.target
X.shape
X
```

```python
y
```

```python
# combient d'exemple de chaque classe4
np.bincount(iris.target)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
X_train
```

```python
#construire le model
modelarbre=DecisionTreeClassifier()
modelarbre.fit(X_train,y_train)
```

```python
#calculer le taux de perfermance
t=modelarbre.score(X_train,y_train)
print("le taux = ",t)
```

```python
#calculer le taux de perfermance
t=modelarbre.score(X_test,y_test)
print("le taux = ",t)
```

```python
# 1,10
#5,10
#6,10
ndepth=6
nsplit=10
modelarbre=DecisionTreeClassifier(max_depth=ndepth,min_samples_split=nsplit)
modelarbre.fit(X_train,y_train)
t=modelarbre.score(X_test,y_test)
print("le taux = ",t)
```
